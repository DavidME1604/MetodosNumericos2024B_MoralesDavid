1. Para cada uno de los siguientes sistemas lineales, obtenga, de ser posible, una soluci√≥n con m√©todos gr√°ficos. 
Explique los resultados desde un punto de vista geom√©trico. 



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las ecuaciones como funciones
def eq1(x1):
    return -x1 / 2  # De x1 + 2x2 = 0, despejamos x2

def eq2(x1):
    return x1  # De x1 - x2 = 0, despejamos x2

# Crear un rango de valores para x1
x1_values = np.linspace(-10, 10, 400)

# Graficar las ecuaciones
plt.figure(figsize=(8, 6))
plt.plot(x1_values, eq1(x1_values), label=r'$x_1 + 2x_2 = 0$', color='blue')
plt.plot(x1_values, eq2(x1_values), label=r'$x_1 - x_2 = 0$', color='red')

# Calcular el punto de intersecci√≥n
A = np.array([[1, 2], [1, -1]])  # Matriz de coeficientes
b = np.array([0, 0])  # Vector de t√©rminos independientes
solution = np.linalg.solve(A, b)  # Resolver el sistema

# Agregar el punto de intersecci√≥n
plt.scatter(solution[0], solution[1], color='black', zorder=5, label='Intersecci√≥n')

# Configurar la gr√°fica
plt.axhline(0, color='black',linewidth=0.5, linestyle='--')
plt.axvline(0, color='black',linewidth=0.5, linestyle='--')
plt.title("Soluci√≥n Gr√°fica del Sistema de Ecuaciones")
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# Mostrar la soluci√≥n
print(f"La soluci√≥n del sistema es: x1 = {solution[0]}, x2 = {solution[1]}")
```

La soluci√≥n es el punto donde ambas l√≠neas se intersectan. Las l√≠neas no son paralelas ni coinciden completamente, por lo que existe una soluci√≥n √∫nica. Este punto es el valor que satisface ambas ecuaciones.

2. Utilice la eliminaci√≥n gaussiana con sustituci√≥n hacia atr√°s y aritm√©tica de redondeo de dos d√≠gitos para resolver los siguientes sistemas lineales. No reordene las ecuaciones. (La soluci√≥n exacta para cada sistema es $x_1=‚àí1, x_2=2,x_3=3$.) 

Definimos el m√©todo gaussian_elimination() que realiza el proceso de eliminaci√≥n gaussian y sustituci√≥n hacia atr√°s.

```python
def gaussian_elimination(A, b, decimals=2):
    """
    Resoluci√≥n de sistemas de ecuaciones lineales usando eliminaci√≥n gaussiana
    y sustituci√≥n hacia atr√°s.
    """
    # Convertir a matriz aumentada
    n = len(b)
    Ab = np.hstack((A, b.reshape(-1, 1)))

    # Eliminaci√≥n gaussiana
    for i in range(n):
        # Pivoteo parcial (si es necesario)
        max_row = np.argmax(np.abs(Ab[i:, i])) + i
        Ab[[i, max_row]] = Ab[[max_row, i]]

        # Normalizar la fila pivote
        Ab[i] = Ab[i] / Ab[i, i]

        # Eliminar variables de filas inferiores
        for j in range(i + 1, n):
            Ab[j] -= Ab[j, i] * Ab[i]

    # Sustituci√≥n hacia atr√°s
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = Ab[i, -1] - np.sum(Ab[i, i + 1:n] * x[i + 1:n])

    # Redondear a los decimales deseados
    return np.round(x, decimals)
```

Luego definimos las matrices de coeficientes y de variables independientes para pasar el m√©todo antes creado.


```python
# Sistema a: Matriz de coeficientes y vector independiente
A_a = np.array([
    [-1, 4, 1],
    [5/3, 2/3, 2/3],
    [2, 1, 4]
])

b_a = np.array([8, 1, 11])

# Sistema b: Matriz de coeficientes y vector independiente
A_b = np.array([
    [4, 2, -1],
    [1/9, 1/9, -1/3],
    [1, 4, 2]
])

b_b = np.array([-5, -1, 9])

# Sistema a
x_a_gauss = gaussian_elimination(A_a, b_a)

# Sistema b
x_b_gauss = gaussian_elimination(A_b, b_b)

print('La soluci√≥n para el primer sistema de ecuaciones es:', x_a_gauss)
print('La soluci√≥n para el segundo sistema de ecuaciones es:', x_b_gauss)
```

3. Utilice el algoritmo de eliminaci√≥n gaussiana para resolver, de ser posible, los siguientes sistemas lineales, y 
determine si se necesitan intercambios de fila: 

- Para resolver este punto, se utiliz√≥ el siguiente m√©todo:

```python
def gaussian_elimination_exchange(A, b):
    """
    Resoluci√≥n de sistemas de ecuaciones lineales usando eliminaci√≥n gaussiana
    y sustituci√≥n hacia atr√°s, indicando solo los intercambios de filas necesarios.
    """
    # Convertir a matriz aumentada
    n = len(b)
    Ab = np.hstack((A, b.reshape(-1, 1)))
    
    # Lista para registrar los intercambios de fila
    row_swaps = []

    # Eliminaci√≥n gaussiana
    for i in range(n):
        # Pivoteo parcial (si es necesario)
        max_row = np.argmax(np.abs(Ab[i:, i])) + i
        if max_row != i:
            Ab[[i, max_row]] = Ab[[max_row, i]]
            row_swaps.append((i, max_row))

        # Normalizar la fila pivote
        Ab[i] = Ab[i] / Ab[i, i]

        # Eliminar variables de filas inferiores
        for j in range(i + 1, n):
            Ab[j] -= Ab[j, i] * Ab[i]

    # Sustituci√≥n hacia atr√°s
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = Ab[i, -1] - np.sum(Ab[i, i + 1:n] * x[i + 1:n])

    # Devolver el resultado y los intercambios realizados
    return x, row_swaps
```

```python
A_a = np.array([
    [1, -1, 3],
    [3, -3, 1],
    [1, 1, 0]
])

b_a = np.array([2, -1, 3.0])

solution, swaps = gaussian_elimination_exchange(A_a, b_a)
print("La soluci√≥n del sistema es: ", solution)
print("Los intercambios de fila fueron: ", swaps)
```

```python
A_b = np.array([
    [2, -1.5, 3],
    [-1, 0, 2],
    [4, -4.5, 5.0]
])

b_b = np.array([1, 3, 1.0])

solution, swaps = gaussian_elimination_exchange(A_b, b_b)
print("La soluci√≥n del sistema es: ", solution)
print("Los intercambios de fila fueron: ", swaps)
```

```python
A_c = np.array([
    [2, 0, 0, 0],
    [1, 1.5, 0, 0],
    [0, -3, 0.5, 0],
    [2, -2, 1, 1]
])

b_c = np.array([3, 4.5, -6.6, 0.8])

solution, swaps = gaussian_elimination_exchange(A_c, b_c)
print("La soluci√≥n del sistema es: ", solution)
print("Los intercambios de fila fueron: ", swaps)
```

```python
A_d = np.array([
    [1, 1, 0, 1],
    [2, 1, -1, 1],
    [4, -1, -2, 2],
    [3, -1, -1, 2]
])

b_d = np.array([2, 1, 0, -3.0])

solution, swaps = gaussian_elimination_exchange(A_d, b_d)
print("La soluci√≥n del sistema es: ", solution)
print("Los intercambios de fila fueron: ", swaps)
```

4. Use el algoritmo de eliminaci√≥n gaussiana y la aritm√©tica computacional de precisi√≥n de 32 bits para resolver los siguientes sistemas lineales.

Para resolver los siguientes literales se desarroll√≥ el sisguiente m√©todo. El cual implemente float32 para satisfacer el requisito propuesto en la orden.

```python
import numpy as np

def gaussian_elimination_32bit(A, b):
    n = len(b)
    # Convertir la matriz a float32 expl√≠citamente
    A = A.astype(np.float32)
    b = b.astype(np.float32)

    # Matriz aumentada
    Ab = np.hstack([A, b.reshape(-1, 1)]).astype(np.float32)

    # Eliminaci√≥n hacia adelante
    for i in range(n):
        # Pivoteo parcial
        max_row = np.argmax(abs(Ab[i:, i])) + i
        Ab[[i, max_row]] = Ab[[max_row, i]]

        # Hacer ceros debajo del pivote
        for j in range(i + 1, n):
            factor = Ab[j, i] / Ab[i, i]
            Ab[j, i:] -= factor * Ab[i, i:]

    # Sustituci√≥n hacia atr√°s
    x = np.zeros(n, dtype=np.float32)
    for i in range(n - 1, -1, -1):
        x[i] = (Ab[i, -1] - np.dot(Ab[i, i + 1:n], x[i + 1:])) / Ab[i, i]

    return x
```

```python
# Sistema a)
A_a = np.array([
    [1/4, 1/5, 1/6],
    [1/3, 1/4, 1/5],
    [1/2, 1, 2]
], dtype=np.float32)

b_a = np.array([9, 8, 8], dtype=np.float32)

x_a = gaussian_elimination_32bit(A_a, b_a)
print("Soluci√≥n del sistema a):", x_a)
```

```python
# Sistema b)
A_b = np.array([
    [3.333, 15920, -10.333],
    [2.222, 16.71, 9.612],
    [1.5611, 5.1791, 1.6852]
], dtype=np.float32)

b_b = np.array([15913, 28.544, 8.4254], dtype=np.float32)

x_b = gaussian_elimination_32bit(A_b, b_b)
print("Soluci√≥n del sistema b):", x_b)
```

```python
# Sistema c)
A_c = np.array([
    [1, 1/2, 1/3, 1/4],
    [1/2, 1/3, 1/4, 1/5],
    [1/3, 1/4, 1/5, 1/6],
    [1/4, 1/5, 1/6, 1/7]
], dtype=np.float32)

b_c = np.array([1/6, 1/7, 1/8, 1/9], dtype=np.float32)

x_c = gaussian_elimination_32bit(A_c, b_c)
print("Soluci√≥n del sistema c):", x_c)
```

```python
# Sistema d)
A_d = np.array([
    [2, 1, -1, 1, -3],
    [1, 0, 2, -1, 1],
    [0, -2, -1, 1, 0],
    [3, 1, -4, 0, 5],
    [1, -1, -1, -1, 0]
], dtype=np.float32)

b_d = np.array([7, 2, -5, 6, -3], dtype=np.float32)

x_d = gaussian_elimination_32bit(A_d, b_d)
print("Soluci√≥n del sistema d):", x_d)
```

7. Repita el ejercicio 4 con el m√©todo Gauss-Jordan.



Para resolver los siguientes literales se desarroll√≥ el sisguiente m√©todo. El cual implemente float32 para satisfacer el requisito propuesto en la orden.

```python
import numpy as np

def gauss_jordan_32bit(matrix):
    """Aplica el m√©todo de Gauss-Jordan con aritm√©tica de 32 bits."""
    matrix = matrix.astype(np.float32)  # Convertir la matriz a float32
    rows, cols = matrix.shape

    for i in range(rows):
        # Escalamiento del pivote a 1
        pivot = matrix[i, i]
        if pivot == 0:
            raise ValueError("El sistema no tiene soluci√≥n √∫nica.")
        matrix[i] = matrix[i] / pivot
        
        # Hacer ceros en la columna del pivote
        for j in range(rows):
            if i != j:
                matrix[j] -= matrix[j, i] * matrix[i]

    return matrix[:, -1]  # Devuelve las soluciones
```

```python
# Sistema a)
A_a = np.array([
    [1/4, 1/5, 1/6, 9],
    [1/3, 1/4, 1/5, 8],
    [1/2, 1, 2, 8]
], dtype=np.float32)

x_a = gauss_jordan_32bit(A_a)
print("Soluci√≥n del sistema a):", x_a)
```

```python
# Sistema b)
A_b = np.array([
    [3.333, 15920, -10.333, 15913],
    [2.222, 16.71, 9.612, 28.544],
    [1.5611, 5.1791, 1.6852, 8.4254]
], dtype=np.float32)

x_b = gauss_jordan_32bit(A_b)
print("Soluci√≥n del sistema b):", x_b)
```
