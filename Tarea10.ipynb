2. Determine cu√°les de las siguientes matrices son no singulares y calcule la inversa de esas matrices:



```python
def is_singular(matriz):
    if (np.linalg.det(matriz) == 0):
        print('La matriz no es singular, por lo que no tiene inversa')
    else:
        print('La matriz es singular y su inversa es: \n', np.linalg.inv(matriz))
```

```python
matriz_a = np.array([[4, 2, 6], [3, 0, 7], [-2, -1, -3]])
is_singular(matriz_a)
```

```python
matriz_b = np.array([[1, 2, 0], [2, 1, -1], [3, 1, 1]])
is_singular(matriz_b)
```

```python
matriz_c = np.array([[1, 1, -1, 1], [1, 2, -4, -2], [2, 1, 1, 5], [-1, 0, -2, -4]])
is_singular(matriz_c)
```

```python
matriz_d = np.array([[4, 0, 0, 0], [6, 7, 0, 0], [9, 11, 1, 0], [5, 4, 1, 1]])
is_singular(matriz_d)
```

4. Encuentre los valores de A que hacen que la siguiente matriz sea singular.



```python
from sympy import Matrix, symbols, solve

alpha = symbols('alpha')

A = Matrix([[1, -1, alpha],
            [2, 2, 1],
            [0, alpha, -3/2]])

det_A = A.det()
print("El determinante de A es:", det_A)

soluciones = solve(det_A, alpha)
print("Valores de alpha para los que la matriz es singular:", soluciones)
```

6. Factorice las siguientes matrices en la descomposici√≥n LU mediante el algoritmo de factorizaci√≥n LU 
con $l_{ii} = 1$ para todas las i.

```python
from scipy.linalg import lu
def descomposicion_LU(A):
    P, L, U = lu(A)

    L = L / np.diag(L)[:, None]

    print("Matriz original A:")
    print(A)

    print("\nMatriz de permutaci√≥n P:")
    print(P)

    print("\nMatriz triangular inferior L (con unos en la diagonal):")
    print(L)

    print("\nMatriz triangular superior U:")
    print(U)
```

```python
A = np.array([
    [2, -1, 1],
    [3, 3, 9],
    [3, 3, 5]
])

descomposicion_LU(A)
```

7. Modifique el algoritmo de eliminaci√≥n gaussiana de tal forma que se pueda utilizar para resolver un sistema lineal usando la descomposici√≥n LU y, a continuaci√≥n, resuelva los siguientes sistemas lineales.

```python
import numpy as np

def eliminacion_gaussiana_LU(A: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    ## Parameters
    ``A``: matriz cuadrada de tama√±o n-by-n.
    
    ## Return
    ``L``: matriz triangular inferior (con unos en la diagonal).
    ``U``: matriz triangular superior.
    """
    A = np.array(A, dtype=float)
    n = A.shape[0]
    L = np.zeros((n, n), dtype=float)

    for i in range(n):
        L[i, i] = 1

        for j in range(i + 1, n):
            m = A[j, i] / A[i, i]
            A[j, i:] -= m * A[i, i:]
            L[j, i] = m

    U = A
    return L, U


def resolver_sistema_LU(L: np.ndarray, U: np.ndarray, b: np.ndarray) -> np.ndarray:
    """Resuelve un sistema lineal dado L, U y el vector b."""
    n = b.shape[0]

    y = np.zeros(n)
    for i in range(n):
        y[i] = b[i] - np.dot(L[i, :i], y[:i])

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i + 1:], x[i + 1:])) / U[i, i]

    return x
```

```python
A = np.array([
    [2, -1, 1],
    [3, 3, 9],
    [3, 3, 5]
])

b = np.array([-1, 0, 4])

L, U = eliminacion_gaussiana_LU(A)
x = resolver_sistema_LU(L, U, b)

print('La solucion del sistema es:', x)
```

```python
A = np.array([
    [1.012, -2.132, 3.104],
    [-2.132, 4.096, -7.013],
    [3.104, -7.013, 0.014]
])

b = np.array([1.984, -5.049, -3.895])

L, U = eliminacion_gaussiana_LU(A)
x = resolver_sistema_LU(L, U, b)

print('La solucion del sistema es:', x)
```
