3. Utilice el método de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

```python
from typing import Callable


def ODE_euler(
    *,
    a: float,
    b: float,
    f: Callable[[float, float], float],
    y_t0: float,
    N: int,
) -> tuple[list[float], list[float], float]:
    """Solves (numerically) an ODE of the form
        dy/dt = f(t, y)
            y(t_0) = y_t0, a <= t_0 <= b
    using the Euler method for the N+1 points in the time range [a, b].

    It generates N+1 mesh points with:
        t_i = a + i*h, h = (a - b) / N,
    where h is the step size.


    ## Parameters
    ``a``: initial time
    ``b``: final time
    ``f``: function of two variables ``t`` and ``y``
    ``y_t0``: initial condition
    ``N``: number of mesh points

    ## Return
    ``ys``: a list of the N+1 approximated values of y
    ``ts``: a list of the N+1 mesh points
    ``h``: the step size h

    """
    h = (b - a) / N
    t = a
    ts = [t]
    ys = [y_t0]

    for _ in range(N):
        y = ys[-1]
        y += h * f(t, y)
        ys.append(y)

        t += h
        ts.append(t)
    return ys, ts, h
```

**a.** 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
    return y/t - (y/t)**2

N = 10 
a, b = 1, 2 
y_t0 = 1  

resultado1 = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
t_euler = resultado1[1]
y_euler = resultado1[0]

for t, y in zip(t_euler, y_euler):
    print(f"t: {t}, y: {y}")

plt.plot(t_euler, y_euler, label="Solución Aproximada N = 10 (Euler)", marker='o', color='r')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solución Exacta vs Aproximada (Método de Euler)')
plt.legend()
plt.grid(True)
plt.show()
```

5. Utilice los resultados del ejercicio 3 y la interpolación lineal para aproximar los siguientes valores de 𝑦(𝑡). Compare 
las aproximaciones asignadas para los valores reales obtenidos mediante las funciones determinadas en el ejercicio 4.


**a.**

```python
t = np.array(resultado1[1])
y = np.array(resultado1[0])

coeficientes = np.polyfit(t, y, 1)

def interpolar(t, coeficientes):
    return np.polyval(coeficientes, t)

t_grafica = np.linspace(min(t), max(t), 100)
y_grafica = interpolar(t_grafica, coeficientes)

def f_real(t):
    return t/(1 + np.log(t))


t_valores = np.array([1.25, 1.93])

y_reales = np.array([f_real(1.25), f_real(1.93)])

y_interpolados = interpolar(t_valores, coeficientes)

print(coeficientes)

errores = abs(y_interpolados - y_reales)

for t_val, y_real, y_interp, error in zip(t_valores, y_reales, y_interpolados, errores):
    print(f"Error en y({t_val}): Interpolado = {y_interp}, Real = {y_real}, Error = {error}")

plt.scatter(t, y, color='red', label='Datos originales')
plt.plot(t_grafica, y_grafica, color='blue', label='Interpolación lineal')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title('Interpolación Lineal')
plt.show()
```
