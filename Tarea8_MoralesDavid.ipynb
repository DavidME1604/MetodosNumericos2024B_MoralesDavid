1. Dados los datos:



**a.** Construya el polinomio por mÃ­nimos cuadrados de grado 1 y calcule el error.  
**b.** Construya el polinomio por mÃ­nimos cuadrados de grado 2 y calcule el error.  
**c.** Construya el polinomio por mÃ­nimos cuadrados de grado 3 y calcule el error.  
**d.** Construya el polinomio por mÃ­nimos cuadrados de la forma $y = b e^{ax}$ y calcule el error.  
**e.** Construya el polinomio por mÃ­nimos cuadrados de la forma $y = b x^a$ y calcule el error.

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Datos proporcionados
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Polinomio de grado 1
coeff_1 = np.polyfit(xi, yi, 1)
poly_1 = np.poly1d(coeff_1)
error_1 = np.sum((yi - poly_1(xi))**2)
print("Literal a)")
print(f"Polinomio de grado 1: y = {coeff_1[0]} * x + {coeff_1[1]}")
print(f"Error cuadrÃ¡tico para el polinomio de grado 1: {error_1}\n")

# Polinomio de grado 2
coeff_2 = np.polyfit(xi, yi, 2)
poly_2 = np.poly1d(coeff_2)
error_2 = np.sum((yi - poly_2(xi))**2)
print("Literal b)")
print(f"Polinomio de grado 2: y = {coeff_2[0]} * x^2 + {coeff_2[1]} * x + {coeff_2[2]}")
print(f"Error cuadrÃ¡tico para el polinomio de grado 2: {error_2}\n")

# Polinomio de grado 3
coeff_3 = np.polyfit(xi, yi, 3)
poly_3 = np.poly1d(coeff_3)
error_3 = np.sum((yi - poly_3(xi))**2)
print("Literal c)")
print(f"Polinomio de grado 3: y = {coeff_3[0]} * x^3 + {coeff_3[1]} * x^2 + {coeff_3[2]} * x + {coeff_3[3]}")
print(f"Error cuadrÃ¡tico para el polinomio de grado 3: {error_3}\n")

# FunciÃ³n exponencial y = b * e^(a * x)
def exp_func(x, a, b):
    return b * np.exp(a * x)

params_exp, _ = curve_fit(exp_func, xi, yi)
a_exp, b_exp = params_exp
error_exp = np.sum((yi - exp_func(xi, a_exp, b_exp))**2)

print("Literal d)")
print(f"FunciÃ³n exponencial: y = {b_exp} * e^({a_exp} * x)")
print(f"Error cuadrÃ¡tico para la funciÃ³n exponencial: {error_exp}\n")

# FunciÃ³n potencia y = b * x^a
def power_func(x, a, b):
    return b * x**a

params_power, _ = curve_fit(power_func, xi, yi)
a_power, b_power = params_power
error_power = np.sum((yi - power_func(xi, a_power, b_power))**2)

print("Literal e)")
print(f"FunciÃ³n potencia: y = {b_power} * x^{a_power}")
print(f"Error cuadrÃ¡tico para la funciÃ³n potencia: {error_power}\n")

# GrÃ¡fica
x_fit = np.linspace(min(xi), max(xi), 100)
plt.figure(figsize=(10, 6))

# Datos originales
plt.scatter(xi, yi, color="black", label="Datos originales", zorder=5)

# Ajustes
plt.plot(x_fit, poly_1(x_fit), label="Polinomio grado 1", linestyle='--')
plt.plot(x_fit, poly_2(x_fit), label="Polinomio grado 2", linestyle='-.')
plt.plot(x_fit, poly_3(x_fit), label="Polinomio grado 3", linestyle=':')
plt.plot(x_fit, exp_func(x_fit, a_exp, b_exp), label=r"$y = be^{ax}$ (Exponencial)", linestyle='-')
plt.plot(x_fit, power_func(x_fit, a_power, b_power), label=r"$y = bx^a$ (Potencial)", linestyle='--')

# PersonalizaciÃ³n
title = "Ajustes por mÃ­nimos cuadrados"
plt.title(title, fontsize=14)
plt.xlabel("$x_i$", fontsize=12)
plt.ylabel("$y_i$", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

3.  La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemÃ¡ticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemÃ¡ticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuaciÃ³n de la recta por mÃ­nimos cuadrados para estos datos. 



```python
import matplotlib.pyplot as plt

# Datos extraÃ­dos de la tabla
act_scores = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 29, 28, 
                       27, 29, 21, 28, 28, 26, 24])
average_scores = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38, 3.53, 2.03, 3.75, 
                           3.65, 3.87, 3.75, 1.66, 3.12, 2.96, 2.92, 2.81])

# Ajuste por mÃ­nimos cuadrados para una lÃ­nea recta
coeff_linear = np.polyfit(act_scores, average_scores, 1)
linear_fit = np.poly1d(coeff_linear)

# EcuaciÃ³n de la recta
m, b = coeff_linear

# Generar puntos para la lÃ­nea ajustada
x_line = np.linspace(min(act_scores), max(act_scores), 100)
y_line = linear_fit(x_line)

# Graficar los datos y la lÃ­nea de ajuste
plt.figure(figsize=(10, 6))
plt.scatter(act_scores, average_scores, color="blue", label="Datos", zorder=3)
plt.plot(x_line, y_line, color="red", label=f"Ajuste: y = {m:.4f}x + {b:.4f}", zorder=2)
plt.title("Ajuste por mÃ­nimos cuadrados: PuntuaciÃ³n ACT vs Promedio de puntos")
plt.xlabel("PuntuaciÃ³n ACT")
plt.ylabel("Promedio de puntos")
plt.legend()
plt.grid(True, zorder=1)
plt.show()

# Imprimir la ecuaciÃ³n de la recta
print(f"EcuaciÃ³n de la recta ajustada: y = {m}x + {b}")
```

El siguiente conjunto de datos, presentado al SubcomitÃ© Antimonopolio del Senado, muestra las caracterÃ­sticas comparativas de supervivencia durante un choque de automÃ³viles de diferentes clases. Encuentre la recta por mÃ­nimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehÃ­culos que participaron en un accidente en los que la lesiÃ³n mÃ¡s grave fue fatal o seria). 

```python
# Datos extraÃ­dos de la tabla
peso_promedio = np.array([4800, 3700, 3400, 2800, 1900])  # en libras
porcentaje_presentacion = np.array([3.1, 4.0, 5.2, 6.4, 9.6])  # en porcentaje

# Ajuste por mÃ­nimos cuadrados para una lÃ­nea recta
coeff_linear_accidentes = np.polyfit(peso_promedio, porcentaje_presentacion, 1)
linear_fit_accidentes = np.poly1d(coeff_linear_accidentes)

# EcuaciÃ³n de la recta
m_acc, b_acc = coeff_linear_accidentes

# Generar puntos para la lÃ­nea ajustada
x_line_acc = np.linspace(min(peso_promedio), max(peso_promedio), 100)
y_line_acc = linear_fit_accidentes(x_line_acc)

# Graficar los datos y la lÃ­nea de ajuste
plt.figure(figsize=(10, 6))
plt.scatter(peso_promedio, porcentaje_presentacion, color="blue", label="Datos", zorder=3)
plt.plot(x_line_acc, y_line_acc, color="red", label=f"Ajuste: y = {m_acc:.4f}x + {b_acc:.4f}", zorder=2)
plt.title("Ajuste por mÃ­nimos cuadrados: Peso promedio vs Porcentaje de presentaciÃ³n")
plt.xlabel("Peso promedio (lb)")
plt.ylabel("Porcentaje de presentaciÃ³n")
plt.legend()
plt.grid(True, zorder=1)
plt.show()

# Imprimir la ecuaciÃ³n de la recta
print(f"EcuaciÃ³n de la recta ajustada: y = {m_acc}x + {b_acc}")
```
