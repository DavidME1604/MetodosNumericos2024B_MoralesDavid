## MÃ©todo de la secante 

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the function
def f(x):
    return x**3 - 3*x**2 + x - 1

# Create an array of x values
x = np.linspace(-2, 4, 400)
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x^3 - 3x^2 + x - 1')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Plot of the function f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.optimize import newton

x0 = 100

x_r = newton(f, x0)

x_r
```

```python
# Generate x values
x = np.linspace(-5, 5, 400)
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2 - 4', color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.scatter(x_r, f(x_r), color='red', zorder=5, label='Roots')
plt.title('Plot of the function and its roots')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-10, 10)
plt.show()
```

## MÃ©todo de Newton 

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y

def f_prime(x, *args):
    return 3*x**2 - 6*x + 1

x0 = 100
points = []
x_r = newton(
    f, x0, fprime=f_prime, args=(points, True), maxiter=100
)


x_points = [point[0] for point in points]
plt.figure(figsize=(8, 6))
plt.plot(x_points, "o-", label="x points")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points in points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```

### No converge $x_0 = 1$

```python
x0 = 1
points = []
x_r, alg_info = newton(
    f,
    x0,
    fprime=f_prime,
    args=(points, False),
    maxiter=50,
    disp=False,  # No mostrar mensajes de error cuando no converge
    full_output=True,  # Return algorithm information
)
print(alg_info)
x_points = [point[0] for point in points]
plt.figure(figsize=(8, 6))
plt.plot(x_points, "o-", label="x points")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points in points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```
