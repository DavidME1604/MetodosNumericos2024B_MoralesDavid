```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

def f(x: float) -> float:
    return x**3 - 3*x**2 + x - 1

def f_prime(x: float) -> float:
    return 3*x**2 - 6*x + 1

def newton_con_estim(func, x0, func_prime, show=False, tol=1e-7, max_iter=100):
    estimaciones = []
    x = x0
    for i in range(max_iter):
        y = func(x)
        estimaciones.append((x, y))
        if show:
            print(f'Iteracin {i}: x = {x}, f(x) = {y}')
        if abs(y) < tol:
            break
        x = x - y / func_prime(x)
    return x, estimaciones

# Initial guess
x0 = 100
raiz, estimaciones = newton_con_estim(f, x0, f_prime, show=True)

# Prepare data for plotting
x_estim, y_estim = zip(*estimaciones)

# Plotting the estimates
plt.plot(x_estim, y_estim, marker='o', color='red', label='Estimaciones de Newton')
plt.axhline(0, color='black', lw=0.5)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.title("Trayectoria de las estimaciones de las races")
plt.grid()
plt.show()

print("\nRaz encontrada:", raiz)
```
